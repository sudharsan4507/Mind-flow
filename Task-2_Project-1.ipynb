```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "Global_Superstore(CSV).csv"
df = pd.read_csv(file_path)
```

```python
# Data Cleaning
# Handle missing values for 'Postal Code' (only column with missing values)
df['Postal Code'].fillna('Unknown')
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Select only numeric columns for outlier detection and statistical analysis
numeric_cols = df.select_dtypes(include=[np.number])

# Detect and handle outliers using IQR for numeric columns
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
outlier_condition = (numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))
df = df[~outlier_condition.any(axis=1)]
```

```python
# Statistical Analysis
summary_stats = numeric_cols.describe()
correlations = numeric_cols.corr()
```

```python
# Data Visualization
# Histogram for Sales
plt.figure(figsize=(8, 5))
sns.histplot(df['Sales'], kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Profit
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Profit'])
plt.title('Profit Boxplot')
plt.xlabel('Profit')
plt.show()
```

```python
# Heatmap for correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
